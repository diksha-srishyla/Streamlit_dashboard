```python
import pandas as pd
```

Reading in the dataset

```python
dataset = pd.read_csv("assignment_dataset.csv")
```

Removing rows missing a risk score or information on "chulha" usage

```python
dataset.dropna(subset = ["q46", "q25"], inplace=True)
```

Filtering data

```python
filtered = dataset[(dataset['q2']=="Yes") & (dataset['q7']>30) & (dataset['q46']>3)]
```

Mapping facility names to labels

```python
filtered.loc[filtered["health_facility"] == "health_facility9", "health_facility"] = "PHC Ramanagar"
filtered.loc[filtered["health_facility"] == "health_facility8", "health_facility"] = "PHC Dhamred"
filtered.loc[filtered["health_facility"] == "health_facility7", "health_facility"] = "PHC Bhanokhar"
filtered.loc[filtered["health_facility"] == "health_facility6", "health_facility"] = "PHC Bahatukala"

filtered.loc[filtered["health_facility"] == "health_facility5", "health_facility"] = "CHC Tahla"
filtered.loc[filtered["health_facility"] == "health_facility4", "health_facility"] = "CHC Rajgarh"
filtered.loc[filtered["health_facility"] == "health_facility3", "health_facility"] = "CHC Pinan"
filtered.loc[filtered["health_facility"] == "health_facility2", "health_facility"] = "CHC Laxmangarh"
filtered.loc[filtered["health_facility"] == "health_facility1", "health_facility"] = "CHC Harsana"
```

Summarizing number screened in each facility/block

```python
screened_count = filtered.groupby(["health_facility","block"]).count()["q46"]
screened_count = pd.DataFrame(screened_count)
screened_count = screened_count.reset_index()

screened_count = screened_count.rename(columns={"q46":"Count", "health_facility":"Health Facility", "block":"Block"})

```

```python
total_screened = screened_count["Count"].sum()

screened_count["Percentage screened"] = ((screened_count["Count"]/total_screened)*100)
screened_count["Percentage screened"] = screened_count["Percentage screened"].round(2)

screened_count.loc['Total'] = screened_count["Count"].sum()

screened_count.at['Total','Health Facility'] = ""
screened_count.at['Total','Block'] = ""
screened_count.at['Total','Percentage screened'] = ""
```

```python
screened_count.to_csv("screened_count.csv")
```

Analyzing mean risk score by block and facility

```python
mean_risk = filtered.groupby(["health_facility", "block"])["q46"].mean().round(2)

mean_risk = pd.DataFrame(mean_risk)
mean_risk = mean_risk.reset_index()

mean_risk = mean_risk.rename(columns={"q46":"Mean risk score", "health_facility":"Health Facility", "block":"Block"})

mean_risk.to_csv("mean_risk.csv")
```

Analyzing by chulha use (yes/no)

```python
chulha_count = filtered.groupby(["health_facility", "block", "q25"]).count()["q46"]

chulha_count = pd.DataFrame(chulha_count)
chulha_count = chulha_count.reset_index()

chulha_count = chulha_count.rename(columns={"q25":"Chulha Use", "q46":"Count", "health_facility":"Health Facility", "block":"Block"})

chulha_count.to_csv("chulha_count.csv")
```
